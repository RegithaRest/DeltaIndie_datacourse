# Praktik Data

```python
data_jualan = pd.read_excel("data_penjualan-Copy.xlsx")
data_jualan
```

```python
data_jualan.isna().sum()
```

```python
#jenis produk---> mode()
#jumlah order --> mean()
#harga --------> mean()
#total --------> jumlah * harga
```

```python
#jenis produk
data_jualan["Jenis Produk"].fillna(data_jualan["Jenis Produk"].mode()[0], inplace=True)
```

```python
#jumlah order
data_jualan["Jumlah Order"].fillna(data_jualan["Jumlah Order"].mean(),inplace=True)
```

```python
#jumlah_harga
data_jualan["Harga"].fillna(data_jualan["Harga"].mean(), inplace=True)
```

```python
#total
data_jualan.loc[data_jualan["Total"].isna(), "Total"] = data_jualan["Jumlah Order"] * data_jualan["Harga"]
```

```python
data_jualan.isna().sum()
```

# Penugasan (jenis produk HVS)
1. Analisa Penjualan setiap bulan dari produk kalian --> Jumlah penjualan setiap bulan

2. Visualisasikan pendapatan tertinggi setiap bulannya. Hitung total pendapatan prduk setiap bulan--> list, {"Januari": 20000; "Februari: 300000}--> visualisasikan

3. kasih legenda setiap visualsiasi

```python
#filter data (HVS)
produk_hvs = data_jualan[data_jualan["Jenis Produk"]=="HVS"] #ini hanya mendapatkan data HVS 
produk_hvs
```

```python
produk_hvs = produk_hvs.copy()
```

```python
produk_hvs["Tanggal"] = pd.to_datetime(produk_hvs["Tanggal"])
```

```python
produk_hvs
```

```python
produk_hvs["Bulan"] = produk_hvs["Tanggal"].dt.month_name()
```

```python
produk_hvs
```

```python
#melakukan penghitungan total jumlah penjualan dan total pendapatan perbulannya
penjualan_bulanan = produk_hvs.groupby("Bulan").agg(
jumlah_penjualan=('Jumlah Order', "sum"),
total_pendapatan = ("Total","sum")
).reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
```

```python
penjualan_bulanan
```

```python
#barchart

#membuat canvas
plt.figure(figsize=(20,10))

#deklarasi colorrs untuk memberikan warna legenda lebih dari 1
colors = plt.cm.tab20.colors

#membuat barchart dengan masukan data x dan y, lalu memberikan label pada data serta menyeting warna setiap kkolom X
plt.bar(penjualan_bulanan.index, penjualan_bulanan["jumlah_penjualan"], label=penjualan_bulanan.index, color=colors[:len(penjualan_bulanan.index)])

#kasih title
plt.title("Jumlah Penjualan HVS Setiap Bulan ( 2 Periode )",fontsize=25)

#label untuk X dan Y
plt.xlabel("BULAN",fontsize=25)
plt.ylabel("JUMLAH ORDER",fontsize=25)

#kasih rotasi (opsional)
plt.xticks(rotation=45)
plt.yticks(rotation=45)

#aktifkan si legend
plt.legend(penjualan_bulanan.index, title="Bulan", loc="upper right")
#plt.grid(True)
plt.show()
```

```python
#per tahun

produk_hvs_tahunan = produk_hvs.copy()
```

```python
produk_hvs_tahunan
```

```python
#convert
produk_hvs_tahunan["Tanggal"] = pd.to_datetime(produk_hvs_tahunan["Tanggal"])
```

```python
produk_hvs_tahunan_2022 = produk_hvs_tahunan[produk_hvs_tahunan["Tanggal"].dt.year==2022]
```

```python
produk_hvs_tahunan_2022
```

```python
total_perbulan_2022 = produk_hvs_tahunan_2022.groupby("Bulan").agg(
jumlah_order_bulan_2022 = ("Jumlah Order","sum"),
jumlah_total_bulan_2022 = ("Total", "sum")
).reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
```

```python
total_perbulan_2022
```

```python
# bikin canvas
plt.figure(figsize=(10,5))

#seting warna
colors = plt.cm.tab20.colors

#setting barnya
plt.bar(total_perbulan_2022.index, total_perbulan_2022["jumlah_order_bulan_2022"],  label=total_perbulan_2022.index, color=colors[:len(total_perbulan_2022.index)])
plt.legend(total_perbulan_2022.index)
plt.title("PENJUALAN PER BULAN DI 2022")
plt.show()
```
