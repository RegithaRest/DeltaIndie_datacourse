```python
filter_data3_dobel= filter_data3_dobel[(filter_data3_dobel["Jenis Produk"]=="Dupleks310") & (filter_data3_dobel["Total"]>100000)]
filter_data3_dobel
```

```python
filter_data3_dobel["Tahun"] = filter_data3_dobel["Tanggal"].dt.year
filter_data3_dobel
```

```python
import matplotlib.pyplot as plt
```

```python
penghasilan2_periode_data3 = filter_data3_dobel.groupby("Bulan").agg(
total_order = ("Jumlah Order","sum"),
total_pendapatan = ("Total","sum")
).reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
```

```python
penghasilan2_periode_data3 
```

```python
#membuat canvas
plt.figure(figsize=(20,10))

#deklarasi colorrs untuk memberikan warna legenda lebih dari 1
colors = plt.cm.tab20.colors

#membuat barchart dengan masukan data x dan y, lalu memberikan label pada data serta menyeting warna setiap kkolom X
plt.bar(penghasilan2_periode_data3 .index, penghasilan2_periode_data3 ["total_order"], label=penghasilan2_periode_data3 .index, color=colors[:len(penghasilan2_periode_data3.index)])

#kasih title
plt.title("Total Order Data 3",fontsize=25)

#label untuk X dan Y
plt.xlabel("BULAN",fontsize=25)
plt.ylabel("Total Order",fontsize=25)

#kasih rotasi (opsional)
plt.xticks(rotation=45)
plt.yticks(rotation=45)

#aktifkan si legend
plt.legend(penghasilan2_periode_data3 .index, title="Bulan", loc="upper right")
#plt.grid(True)
plt.show()
```
